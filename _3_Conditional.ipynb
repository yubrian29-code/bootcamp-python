```python
# elif

# score -> grade
# >= 90 -> A
# >= 80 -> B
# Otherwise F
score = 92
if (score > 90):
  grade = 'A'
elif (score >= 80):
  grade = 'B'
else:
  grade = 'F'
  print(f'grade={grade}')
```

```python
# Java: equals
a = "hello"
b = "hello"
c = "".join(['h','e','l','l','o'])

print(type(b))
print(type(c))

print(a==b) # same value
print(a==c) # same value
print(a is b)
print(a is c)
```
