```python
from datetime import datetime, date, timedelta
from dateutil.relativedelta import relativedelta
```

```python
print(datetime.now())
print(datetime.today())

birthday = date(2020, 12, 31)
print(birthday)

# add days
print(birthday + timedelta(days=10))
```

```python
join_date = date.today()
probation_end_date = join_date + relativedelta(month=3) - relativedelta(days=1)
print(probation_end_date)

# years, month, days
```

```python
formatted_date = datetime.now().strftime("%Y-%m-%d %H:%M:%S")
print(type(formatted_date))
print(formatted_date)

str_datetime = "2025/12/31 23:59:59"
special_datetime = datetime.strptime(str_datetime, "%Y/%m/%d %H:%M:%S")
print(type(special_datetime))
print(special_datetime)
```
