```python
# Data type
# int, float, str, boolean
x = 3
print(type(x)) # Java: x.getClass(), instanceOf()
print(x)

x = 3.3
print(type(x)) # Variable Type Dynamic
print(x)

x = "hello"
print(type(x))
print(x)

x = False
print(type(x))
print(x)
```

```python
y = "hello"
print(y * 3)

# Python: Str + other type -> error
# y = y + 2
y = y + str(2)
print(y)

a = 10
# ++a
a += 1
print(a)

b = 3.5
c = 10
v = b + c
print(type(v)) # float

j = 3.5
k = 2
o = j * k
print(type(o)) # float

print(True + 2) # True 1
print(False + 2) # False 0
print(True + 2.5)
```
