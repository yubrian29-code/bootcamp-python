```python
# list
languages = ['Java', 'Python', 'C++']

languages.append('C#')
print(languages)

# remove, insert, sort, pop, clear, len
# list1 + list2

languages.remove('Python') # remove first, then break
print(languages)

languages.insert(3, "JavaScript")
print(languages)

languages.sort()
print(languages)

pop_result = languages.pop(2)
print(pop_result)

print(len(languages))

human_languages = ['English', 'Chinese']
languages = human_languages + languages
print(languages)

languages.clear
```

```python
# dictionaty (Java: map)

person = {
  "name" : "John",
  "age" : 17
}

print(type(person))
print(person['name'])
print(person['age'])

trash = {
  17 : "Steven",
  0.99 : "Trump",
  True : False
}

print(trash[0.99])

print(trash.keys())
print(trash.values())

for k,v in trash.items():
  print(f"key={k},value={v}")

person['age'] = 100
print(person)

person['age'] = [100, 200, 300]
print(person)
```
