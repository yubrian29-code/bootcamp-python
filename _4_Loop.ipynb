```python
names = ['John', 'Steven', 'Jennie']

for name in names:
  print(name)

for i in range(3):
  print(names[i])

for ch in "Java":
  print(ch) # str
  print(type(ch))

word = "hello world"
# check if o exists -> boolean
o_exists = False
for c in word:
  if (c == 'o'):
    o_exists = True
    break
print(o_exists)

print('o' in word)
```

```python
numbers =[1, 100, -6, 20000]
print(max(numbers))
print(min(numbers))
print(sum(numbers))
print(len(numbers))

# indexOf
print(numbers.index(20000)) # 3
print(numbers.index(max(numbers))) # 3

print(sum(numbers) / len(numbers))
```
