```python
# String -> char[]

word = 'Python'
print(word[0])
# print(word[6]) # error
print(word[-1]) # Java: length - 1
print(word[-3])

print(word[:3])
print(word[1:3])
print(word[2:])

print(word.upper())
print(word.lower())
print(word.replace('t', 'x'))
```
